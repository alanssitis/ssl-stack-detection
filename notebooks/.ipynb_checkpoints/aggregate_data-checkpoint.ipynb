```python
import pandas as pd
```

```python
go_data = pd.read_csv("../data/2023_12_13_go_client_go_server_p10000.csv")
openssl_1_0_2_data = pd.read_csv("../data/2023_12_13_openssl_1_0_2_client_go_server_p10000.csv")
openssl_1_1_1_data = pd.read_csv("../data/2023_12_13_openssl_1_1_1_client_go_server_p10000.csv")
openssl_3_1_1_data = pd.read_csv("../data/2023_12_13_openssl_3_1_1_client_go_server_p10000.csv")
openssl_3_2_0_data = pd.read_csv("../data/2023_12_13_openssl_3_2_0_client_go_server_p10000.csv")
rustls_21_data = pd.read_csv("../data/2023_12_13_rustls_21_client_go_server_p10000.csv")
rustls_22_data = pd.read_csv("../data/2023_12_13_rustls_22_client_go_server_p10000.csv")
```

```python
openssl_old_rename = {
    "AES128-SHA256": "TLS_RSA_WITH_AES_128_CBC_SHA256",
    "AES128-GCM-SHA256": "TLS_RSA_WITH_AES_128_GCM_SHA256",
    "AES256-GCM-SHA384": "TLS_RSA_WITH_AES_256_GCM_SHA384",
    "ECDHE-RSA-AES128-SHA256": "TLS_ECDHE_RSA_WITH_AES_128_CBC_SHA256",
    "ECDHE-RSA-AES128-GCM-SHA256": "TLS_ECDHE_RSA_WITH_AES_128_GCM_SHA256",
    "ECDHE-RSA-AES256-GCM-SHA384": "TLS_ECDHE_RSA_WITH_AES_256_GCM_SHA384",
}
openssl_new_rename = {
    "AES128-SHA256": "TLS_RSA_WITH_AES_128_CBC_SHA256",
    "AES128-GCM-SHA256": "TLS_RSA_WITH_AES_128_GCM_SHA256",
    "AES256-GCM-SHA384": "TLS_RSA_WITH_AES_256_GCM_SHA384",
    "ECDHE-RSA-AES128-SHA256": "TLS_ECDHE_RSA_WITH_AES_128_CBC_SHA256",
    "ECDHE-RSA-AES128-GCM-SHA256": "TLS_ECDHE_RSA_WITH_AES_128_GCM_SHA256",
    "ECDHE-RSA-AES256-GCM-SHA384": "TLS_ECDHE_RSA_WITH_AES_256_GCM_SHA384",
    "ECDHE-RSA-CHACHA20-POLY1305": "TLS_ECDHE_RSA_WITH_CHACHA20_POLY1305_SHA256",
}
rustls_rename = {
    "TLS13_AES_128_GCM_SHA256": "TLS_AES_128_GCM_SHA256",
    "TLS13_AES_256_GCM_SHA384": "TLS_AES_256_GCM_SHA384",
    "TLS13_CHACHA20_POLY1305_SHA256": "TLS_CHACHA20_POLY1305_SHA256",
}

openssl_1_0_2_data = openssl_1_0_2_data.rename(columns=openssl_old_rename)
openssl_1_1_1_data = openssl_1_1_1_data.rename(columns=openssl_new_rename)
openssl_3_1_1_data = openssl_3_1_1_data.rename(columns=openssl_new_rename)
openssl_3_2_0_data = openssl_3_2_0_data.rename(columns=openssl_new_rename)
rustls_21_data = rustls_21_data.rename(columns=rustls_rename)
rustls_22_data = rustls_22_data.rename(columns=rustls_rename)
```

```python
go_data["stack"] = "go"
openssl_1_0_2_data["stack"] = "openssl_1_0_2"
openssl_1_1_1_data["stack"] = "openssl_1_1_1"
openssl_3_1_1_data["stack"] = "openssl_3_1_1"
openssl_3_2_0_data["stack"] = "openssl_3_2_0"
rustls_21_data["stack"] = "rustls_0_21_7"
rustls_22_data["stack"] = "rustls_0_22_0"
```

```python
data = [go_data, openssl_1_0_2_data, openssl_1_1_1_data, openssl_3_1_1_data, openssl_3_2_0_data, rustls_21_data, rustls_22_data]
data = [d.drop(columns=["idx"]) for d in data]
data[0]
```

```python
all_data = pd.concat(data, ignore_index=True)
all_data
```

```python
all_data.to_csv("../data/2023_12_13_all_clients_go_server_p70000.csv", index=False)
```
